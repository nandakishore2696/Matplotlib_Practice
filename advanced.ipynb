```python
batters = pd.read_csv('batter.csv')
batters
```

```python
sample_df = batters.loc[:100].sample(25)
sample_df
```

## Annotations

```python
plt.figure(figsize=(10,7))
plt.scatter(sample_df['avg'], sample_df['strike_rate'] , s=sample_df['runs']/10)

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
sample_df.values
```

```python
# Horizontal and Vertical lines

plt.figure(figsize=(10,7))
plt.scatter(sample_df['avg'], sample_df['strike_rate'] , s=sample_df['runs']/10)

plt.axhline(130, color = 'red')
plt.axhline(140, color = 'green')
plt.axvline(30, color = 'red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])

plt.xlabel('Average score')
plt.ylabel('Strike Rate')
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (15,6))

ax.scatter(batters['avg'], batters['strike_rate'], color = 'red', marker ='*')
ax.set_title("something")
ax.set_xlabel('Average')
ax.set_ylabel('Strike rate')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1, figsize = (10,7))

ax[0].scatter(batters['avg'], batters['strike_rate'])
ax[0].set_title("Avg vs Strike rate")
ax[0].set_xlabel('Average score')
ax[0].set_ylabel('Strike Rate')
ax[1].scatter(batters['avg'], batters['runs'])
ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel('Total score')
ax[1].set_ylabel('Total Rate')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2, figsize = (10,10))

ax[0,0].scatter(batters['avg'], batters['strike_rate'], color = 'red')

ax[0,1].scatter(batters['avg'], batters['runs'])

ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['strike_rate'])
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax1.scatter(batters['avg'], batters['strike_rate'])

ax2 = fig.add_subplot(2,1,2)
ax2.scatter(batters['avg'], batters['runs'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'], batters['strike_rate'])

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'], batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])

ax4 = fig.add_subplot(2,2,4)
ax4.hist(batters['strike_rate'])
```
