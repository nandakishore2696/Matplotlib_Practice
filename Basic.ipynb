1. Set Custom Tick Locations and Labels:
* Syntax: plt.xticks(locations, labels)
2. Adjust Tick Rotation:
* Syntax: plt.xticks(rotation='vertical')
3. Control Tick Frequency:
* Syntax: plt.xticks(step=2), which would display every 2nd tick.
4. Set Tick Font Properties:You can specify font properties such as font size, font weight, font family, etc., for the tick labels.
* Syntax: plt.xticks(fontsize=12, fontweight='bold')
5. Customize Tick Labels:You can use LaTeX expressions or special characters in tick labels.
* Syntax: plt.xticks([0, 1, 2], ['$\pi$', '$2\pi$', '$3\pi$'])
6. Hide Ticks:You can hide ticks from the plot using an empty list [] for locations.
* Syntax: plt.xticks([])
7. Set Minor Ticks:You can set minor ticks to appear on the x-axis.
* Syntax: plt.minorticks_on()
8. Set Tick Visibility:You can control the visibility of ticks using the visible parameter.
* Syntax: plt.xticks(visible=False)

### Scatter can also be used for `MultiVariate Analysis`
* By Specifying `c` - color
* `s` - size

```python
# `c` - color
plt.scatter(tips['total_bill'], tips['tip'], c=tips['sex'].map({'Male': 'blue', 'Female':'red'}))
plt.show()
```

```python
male_data = tips[tips['sex'] == 'Male']
female_data = tips[tips['sex'] == 'Female']
plt.scatter(male_data['total_bill'], male_data['tip'], c='blue', label='Male')
plt.scatter(female_data['total_bill'], female_data['tip'], c='red', label='Female')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.legend(title='Sex')
plt.show()
```

```python
Non_Smoker_data = tips[tips['smoker'] == 'No']
Smoker_data = tips[tips['smoker'] == 'Yes']
plt.scatter(Non_Smoker_data['total_bill'], Non_Smoker_data['tip'], c='blue', label='Non Smoker')
plt.scatter(Smoker_data['total_bill'], Smoker_data['tip'], c='red', label='Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.legend(title='Yes/No')
plt.show()
```

```python
# 's' - size
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*15, color= 'green')
plt.show()
```

## Bar Plot
* Bivariate Analysis
* grouping of categories column by count,sum,...
* Categorical -> Numerical
* Use Case - Aggregate Analysis of groups

```python
children = [10,20,40,10,30]
colors = ['red', 'blue','green','yellow','pink']
plt.bar(colors,children)
plt.show()
```

```python
# color - hidden parameter
children = [10,20,40,10,30]
colors = ['red', 'blue','green','yellow','pink']
plt.bar(colors,children,color='green')
plt.show()
```

```python
# different colors to different for various bars
children = [10,20,40,10,30]
colors = ['red', 'blue','green','yellow','pink']
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors,children,color=colors)
plt.show()
```

```python
# horizontal bars
# if number of categories increases, horizontal is good
children = [10,20,40,10,30]
colors = ['red', 'blue','green','yellow','pink']
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.barh(colors,children,color=colors)
plt.show()
```

```python
df = pd.read_csv("batsman_season_record.csv")
df
```

```python
# side by side bar charts
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2, label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, label='2016')
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2, label='2017')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Plot')
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width=0.3, label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.3, label='2016')
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width=0.3, label='2017')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Plot')
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])-0.25, df['2015'], width=0.25, label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.25, label='2016')
plt.bar(np.arange(df.shape[0])+0.25, df['2017'], width=0.25, label='2017')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Plot')
plt.legend()
plt.show()
```

```python
# Stacked Bar Chart
plt.bar(df['batsman'],df['2015'], label = '2015')
plt.bar(df['batsman'],df['2016'], bottom=df['2015'], label = '2016')
plt.bar(df['batsman'],df['2017'], bottom=(df['2015']+df['2016']), label = '2017')
plt.legend()
plt.show()
```
